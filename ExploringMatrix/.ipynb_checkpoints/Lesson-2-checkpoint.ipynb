## Simulate with scipy's odeint algorithm

### Odeint
`odeint` is a function from the `scipy.integrate` module of the SciPy library, a popular Python library for scientific computing. The `odeint` function is used to solve ordinary differential equations (ODEs) of the form:

$$
\frac{dy}{dt} = f(y, t, \ldots)
$$

- \(y\) is the state vector (which can be multi-dimensional), 
- \(t\) is the independent variable (often time), and 
- \(f\) is a function that specifies the derivatives of the system.

The syntax for `odeint` is:

```python
scipy.integrate.odeint(func, y0, t, args=(), ...)
```

Here's a breakdown of the parameters:

- `func`: This is a function that returns the derivatives of the system to be solved. It must take at least two inputs: the current value of \(y\) and the current value of \(t\). It can also take additional parameters, which are passed using the `args` parameter.
- `y0`: An array specifying the initial state of the system.
- `t`: An array of points at which to solve for \(y\), with the initial condition point as the first element of this sequence.
- `args`: Optional. A tuple containing any extra arguments to pass to `func`.

And it returns:

- An array of the same shape as `y0`, with each row corresponding to the solution at each time point in `t`.

````python
import numpy as np                  # This is for arrays 
from scipy.integrate import odeint  # this is for numerical integration
import matplotlib.pyplot as plt     # this is for plotting
````

````python
def population_growth2(P, t, r):
    dPdt = r * P
    return dPdt
````

````python
# simulaion parameters
t_time = 20
n_points = 100
t = np.linspace(0, t_time, n_points)

# initial conditions
P0  = 10
r = 0.1

# run the simulation
solution = odeint(population_growth2, P0, t, args=(r,))
````

````python
solution = odeint(population_growth2, P0, t, args=(r,))
````

````python
#plot 
plt.plot(t,solution)
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.title('Population Growth Over Time')
````

## Lotka-Volterra 2D model (Predator-Prey model)

In our previous lessons, we examined how the population of a single species can change over time. However, in most ecosystems, species do not exist in isolation; rather, they interact with one another. These interactions are crucial to understanding how populations evolve within a community. To explore such complexities, we can model a two-dimensional system. 

One simple yet illustrative example is the `Lotka-volterra - predator-prey model`. This model considers two types of organisms: predators and prey. It examines how the populations of both groups fluctuate over time due to their interdependent relationships.

### Predator-prey equations:

1. Prey population growth:

$$ \frac{dP}{dt} = \alpha P - \beta PV $$

2. Predator populaton growth
$$ \frac{dV}{dt} = \delta PV - \gamma V $$

- $P$ is the prey population
- $V$ is the predator population 
- $\alpha$ is the growth rate of the prey
- $\beta$ is the predation rate on the prey
- $\delta$ is the efficiency of converting prey to predator
- $\gamma$ is death rate of predators

````python
# lotka-volterra
def lotka_volterra(y, t, alpha, beta, delta, gamma):
    P, V = y
    dPdt = alpha * P - beta * P * V
    dVdt = delta * P * V - gamma * V
    return dPdt, dVdt
````

````python
total_time = 40
n_points = 1001

# initial conditions and params
P0 = 10 # initial prey population
V0 = 5 # initial predator Population 
alpha = 1.0 # Prey growth
beta = 1.0 # Predation rate
delta = 1.0 # predator efficiency
gamma = 2.0 # predator death rate

# odeint inputs
y0 = [P0, V0]
t= np.linspace(0, total_time, n_points) # the time points

# simulate
solution = odeint(lotka_volterra, y0, t, args=(alpha, beta, delta, gamma))
````

````python
# split out the individual variables
P_soln = solution[:,0]
V_soln = solution[:,1]
````

````python
solution = odeint(lotka_volterra, y0, t, args=(alpha, beta, delta, gamma))
````

````python
# plot the results
plt.plot(t, P_soln, label='Prey')
plt.plot(t, V_soln, label = 'Predators')
plt.xlabel('Time')
plt.ylabel('Population Size')
plt.title('Lotka-Volterra Predator-Prey Model')
plt.legend()
````

## Plot vector field phase diagram
A vector field phase diagram for a predator-prey model is a graphical representation that allow us to visualize the dynamics of two interacting populations within an ecosystem: one of predators and the other of prey. 
The vector field phase diagram is plotted in a two-dimensional space, where each axis represents the population size of one of the species. <br>
In this diagram, each point in the phase space represents a possible state of the ecosystem (i.e., a specific number of predators and prey). Vectors (arrows) are drawn at each point, indicating the direction in which the system will evolve from that point. The length and direction of each vector represent the rate of change of the populations. Thus, the vector field shows how the predator and prey populations change over time

### Important functions to plot a vector field
The `np.linspace` function is used to create evenly spaced numbers 
over a specified interval (start, end, number to generate)


The `np.meshgrid` function in NumPy is used to create a rectangular grid out 
of two given one-dimensional arrays representing the Cartesian indexing or Matrix indexing. 

The `quiver` function in matplotlib is typically used to plot vectors as arrows on a 2D or 3D plot, which visually represent vector fields. For 2D vector fields, this method helps to show the size and direction of vectors in 
the domain


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import ipywidgets as widgets
from IPython.display import display
````

````python
# set the max population
x_lim = 15 # The limit for species 1
y_lim = 15 # The limit for species 2

# Generate random predator and prey populations
P_values = np.linspace(0,x_lim,20)
V_values = np.linspace(0,y_lim,20)

PP, VV = np.meshgrid(P_values, V_values)
dP, dV = lotka_volterra([PP, VV],  0, alpha, beta, delta, gamma)
````

````python
# plot vector space
fig, ax = plt.subplots()  # Create a figure and axis
ax.quiver(PP, VV, dP, dV, color='blue')
ax.set_xlabel('Prey population')
ax.set_ylabel('predator population')
````

### Overlay a trajectory

````python
t = np.linspace(0, 15, 1000)  # Time range for the simulation
initial_conditions = [10, 5]  # Example initial conditions
trajectory = odeint(lotka_volterra, initial_conditions, t, args=(alpha, beta, delta, gamma))
````

````python
fig, ax = plt.subplots()  # Create a figure and axis
ax.quiver(PP, VV, dP, dV, color='blue')
ax.plot(trajectory[:, 0], trajectory[:, 1], 'r-')  # 'r-' for a red line
ax.set_xlabel('Prey Population (P)')
ax.set_ylabel('Predator Population (V)')
ax.set_title('Lotka-Volterra Predator-Prey Model with Trajectory')
ax.set_xlim([0, x_lim])
ax.set_ylim([0, y_lim])
plt.show()
````

### Interactive vector field
Widgets refer to interactive GUI elements that allow users to interact with the notebook and manipulate code in real-time. 

Here, we employ it to allow us manipulate model parameters in real time.


````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import ipywidgets as widgets
from IPython.display import display
````

````python
def plot_lotka_volterra(alpha=1.1, beta=0.4, delta=0.1, gamma=0.4, P0=40, V0=9, total_time=15):
    t = np.linspace(0, total_time, 10000) 
    initial_conditions = [P0, V0]
    trajectory = odeint(lotka_volterra, initial_conditions, t, args=(alpha, beta, delta, gamma))
    max_P = max(trajectory[:,0])
    max_V = max(trajectory[:,1])

    P_values = np.linspace(0, max_P, 20)
    V_values = np.linspace(0, max_V, 20)
    PP, VV = np.meshgrid(P_values, V_values)
    dP, dV = lotka_volterra([PP, VV], 0, alpha, beta, delta, gamma)

    fig, ax = plt.subplots()
    ax.quiver(PP, VV, dP, dV, color='blue')
    ax.plot(trajectory[:, 0], trajectory[:, 1], 'r-')
    ax.set_xlim([-0.1 * max_P, max_P * 1.1])
    ax.set_ylim([-0.1 * max_V, max_V * 1.1])
    ax.set_xlabel('Prey Population (P)')
    ax.set_ylabel('Predator Population (V)')
    ax.set_title('Lotka-Volterra Model')
    plt.show()
````

````python
widget = widgets.interactive(plot_lotka_volterra, 
                    alpha=widgets.FloatSlider(value=1.0, min=0, max=10, step=1),
                    beta=widgets.FloatSlider(value=1.0, min=0, max=10, step=1),
                    delta=widgets.FloatSlider(value=0.1, min=0, max=2, step=0.1),
                    gamma=widgets.FloatSlider(value=0.4, min=0, max=2, step=0.1),
                    P0=widgets.IntSlider(value=40, min=0, max=100, step=1),
                    V0=widgets.IntSlider(value=9, min=0, max=100, step=1),
                    total_time=widgets.IntSlider(value=15, min=0, max=100, step=1))

display(widget)
````
